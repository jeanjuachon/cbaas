```python
#The methodology for the Poisson distribution here may not be correct, no assumption checking/transformation done, etc. the purpose is to only build a dummy model
from sklearn.linear_model import PoissonRegressor
y = df2['DEMAND ']
X = df2[["Advertising Expenses (SMS)", "Advertising Expenses(Newspaper ads)", "Advertising Expenses(TV)", "Advertising Expenses(Internet)"]]

# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# scaled = scaler.fit_transform(X)
# X_scaled = pd.DataFrame(scaled)
# X_scaled.columns = ["Advertising Expenses (SMS)", "Advertising Expenses(Newspaper ads)", "Advertising Expenses(TV)", "Advertising Expenses(Internet)"]
```

```python
import statsmodels.api as sm
poisson_training_results = sm.GLM(y, X, family=sm.families.Poisson()).fit()
```

```python
print(poisson_training_results.summary())
```

```python
Assuming that the model is a good fit, we can now interpret the Beta coefficients of the Poisson Regression model to explain the demand, such as:<br>
by holding other coefficients constant, we can a 1 unit increase for SMS advertising expense decreases the expected demand will increase by -.0451 unit.<br>
Also by holding other coefficients, we can expect a 0.57 increase in demand if we increase the newspaper advertising by 1 unit.
```
